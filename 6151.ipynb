INTRODUCTION
In this Dual citizenship dataset information about the Maastricht Center for Citizenship, Migration and Development’s Global Expatriate Dual Citizenship Dataset tracks how 200 countries have, each year since 1960, treated this situation. The extensive documentation provides links to the relevant laws, and descriptions of how each country’s rules have changed.

In that project dealing with EDA (Exploratory Data Analysis) and visualization of MACIMIDE Global expatriate dual Citizenship Dataset.

In this EDA perform some concepts like duplicates values handling, null value handling, describing the data, etc.
Dataset has 10 columns and 10030 rows in total. Each row giving valuable insights in analysis No. of persons around the world with dual or more nationalities.  

Dataset provides 10 columns like ISO, country, country_code, dependency_code, world_region, Year, Dualcit_cat, Dualcit_grouped, Dualcit_binary. In this different columns have different data types like int, floats, and objects. And also perform an encoding concept.
Using this data perform some visualization for essay to understand and analysis the data from the dataset like box, line, bar, scatter, heatmap, pairplot, countplot, groupby bar plot, pieplot, hist plot, dist plot,


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
plt.rcParams["figure.figsize"]=[13,9]
df=pd.read_csv("/MACIMIDE Global Expatriate Dual Citizenship Dataset_V5.00.csv")
df.head()
```

```python
pd.pivot_table(df, values='world_region',index='Year', columns='country', aggfunc='mean')
df.sort_values('world_region', ascending=False)
```

```python
dual_citizenship_data = df['world_region']

mean = dual_citizenship_data.mean()
median = dual_citizenship_data.median()
mode = dual_citizenship_data.mode()[0]
std_dev = dual_citizenship_data.std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")

plt.figure(figsize=(10, 6))
sns.countplot(x=df['world_region'])
plt.title("Global Dual Citizenship Count Plot")
plt.xlabel("world region")
plt.ylabel("Count")
plt.show()
```

Count Plot :
countplot() method is used to Show the counts of observations in
                              each categorical bin using bars. Here it counts all regions using
                              world_region variable. And that 1.0 has highest counting in the
         world_region


```python
plt.figure(figsize=(10, 6))
sns.histplot(x='world_region',color='skyblue', edgecolor='black',data=df)
plt.title("Global Dual Citizenship")
plt.xlabel('world_region')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='world_region', y='country_code', data=df, color='skyblue')
plt.title("Scatter plot")
plt.show()
```

Scatter Plot :: Scatter plots are used to plot data points on horizontal and vertical axis in
 the attempt to show how much one variable is affected by another.


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='country',y='Dualcit_binary', data=df,order=df.country.value_counts().iloc[:5].index,palette='Set2')
plt.title("Comparision between Dualcit_binary and country")
plt.xlabel("Country")
plt.ylabel("Dualcit_binary")
plt.show()
```

```python
variance = df.var()
print("Variance:")
print(variance)

# Calculate the covariance matrix
covariance_matrix = df.cov()
print("\nCovariance Matrix:")
print(covariance_matrix)

```

```python
country_United_States_of_America = df[df['country'] == 'United States of America']
country_Sri_Lanka = df[df['country'] == 'Sri Lanka']
country_Zimbabwe = df[df['country'] == 'Zimbabwe']
country_India = df[df['country'] == 'India']
print("United States of America:")
print(country_United_States_of_America)

print("\nSri Lanka:")
print(country_Sri_Lanka.head())
print("\nZimbabwe:")
print(country_Zimbabwe.head())

print("\nIndia:")
print(country_India.head())
```

```python
dual_citizenship_data= country_India['Dualcit_binary']
plt.figure(figsize=(8, 8))
plt.pie(dual_citizenship_data, labels=dual_citizenship_data, autopct='%1.1f%%', startangle=140)
plt.title('Global Dual Citizenship Analysis')
plt.axis('equal')
plt.show()
```
